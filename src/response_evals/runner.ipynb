# Run the LLM judge

Judges one or more `data/model_responses/<folder>/responses.jsonl` folders on
all 7 dimensions plus a language check (english/filipino/other), with a live
tqdm progress bar (rows complete as their 8 judge calls finish), and writes
`data/persuasion_result/<folder>.csv` + `<folder>_reasoning.jsonl`.

```python
import os
os.environ["KMP_DUPLICATE_LIB_OK"] = "TRUE"
os.environ["OMP_NUM_THREADS"] = "1"
```

### Folders to judge

Edit this list to whatever folders under `data/model_responses/` you want to
run. Each row makes 7 judge calls (one per dimension), so e.g. 1500 rows is
10,500 calls for that folder.

```python
FOLDERS = ["en-eval-on-tl-steer", "tl-eval-on-en-steer"]
CONCURRENCY = 10  # rows judged in parallel; each row fires 8 concurrent calls (7 dims + language)
```

```python
results = {}
for folder in FOLDERS:
    print(f"=== {folder} ===")
    df = await judge_folder(folder, concurrency=CONCURRENCY, show_progress=True)
    results[folder] = df
```

### Quick sanity check on the results

```python
for folder, df in results.items():
    print(f"=== {folder} ===")
    if df is None:
        print("  skipped (missing/empty responses.jsonl)")
        continue
    print(f"  rows: {len(df)}")
    display(df.head())
```
